```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 加载tokenizer和模型
tokenizer = AutoTokenizer.from_pretrained("defog/sqlcoder-7b-2")
model = AutoModelForCausalLM.from_pretrained("defog/sqlcoder-7b-2")

# 设置设备为GPU（如果可用）或CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# 准备输入文本
input_text = "List all customers who have made more than 5 orders."
input_text = "列出所有订单超过5个的客户。"

# 对输入文本进行tokenization
input_ids = tokenizer.encode(input_text, return_tensors="pt").to(device)

# 生成SQL查询
with torch.no_grad():
    output_ids = model.generate(input_ids, max_length=100, num_return_sequences=1)

# 解码生成的SQL查询
generated_sql = tokenizer.decode(output_ids[0], skip_special_tokens=True)

print("Generated SQL Query:", generated_sql)
```

# 自定义pipeline

```python
from transformers import Pipeline


class MyPipeline(Pipeline):
    def _sanitize_parameters(self, **kwargs):
        preprocess_kwargs = {}
        if "maybe_arg" in kwargs:
            preprocess_kwargs["maybe_arg"] = kwargs["maybe_arg"]
        return preprocess_kwargs, {}, {}

    def preprocess(self, inputs, maybe_arg=2):
        model_input = Tensor(inputs["input_ids"])
        return {"model_input": model_input}

    def _forward(self, model_inputs):
        # model_inputs == {"model_input": model_input}
        outputs = self.model(**model_inputs)
        # Maybe {"logits": Tensor(...)}
        return outputs

    def postprocess(self, model_outputs):
        best_class = model_outputs["logits"].softmax(-1)
        return best_class
    
# _sanitize_parameters：提取并清理参数。
# preprocess：将输入数据转换为模型接受的格式。
# _forward：执行模型的前向传播。
# postprocess：对模型的输出进行处理，生成最终结果。
```

_sanitize_parameters 方法用于清理和提取传递给管道的参数。它将可能存在的参数 maybe_arg 提取到 preprocess_kwargs 字典中，并返回三个字典（分别用于预处理、前向传播和后处理）。在这个示例中，只有预处理参数被提取。

preprocess 方法用于将输入数据转换为模型可以接受的格式。在这个示例中，它将输入数据中的 input_ids 转换为 Tensor 并返回一个字典，包含键 model_input。

_forward 方法用于执行模型的前向传播。在这个示例中，它将 model_inputs（一个包含 model_input 键的字典）传递给模型，并返回模型的输出（例如包含 logits 的字典）。_forward 方法用于执行模型的前向传播。在这个示例中，它将 model_inputs（一个包含 model_input 键的字典）传递给模型，并返回模型的输出（例如包含 logits 的字典）。

postprocess 方法用于处理模型输出，生成最终的输出结果。在这个示例中，它对模型的 logits 应用 softmax，并返回结果。
